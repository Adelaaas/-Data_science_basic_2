```
a = ['Gender', 'Ever_Married', 'Graduated', 'Profession', 'Spending_Score', 'Var_1']
train = pd.get_dummies(train, columns=a)
train
```

```
# знаем что все два уникальных значения
# создадим свой словарь исходя из этого
d = {'A': 0, 'B': 1, 'C' : 2, 'D' : 3}

# Заменим категориальные значения в столбце на числа
train['Segmentation'] = train['Segmentation'].map(d)

train.head()
```

```
print(train.isna().sum())
```

```
from sklearn.model_selection import train_test_split

# train - на тренировочные и тестовые
X = train.drop(['Segmentation'], axis=1)
y = train['Segmentation']

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3)

model = LogisticRegression()

# модель учится, для этого ей нужны правильные
model.fit(X_train, y_train)

# тестируем, для признаков у которых нет ответов, мы предсказывает ответ
answers_pred = model.predict(X_test)
```

```
from sklearn.metrics import accuracy_score

# Как оценить насколько хорошо работает ваша модель?

# answers_pred - ответы которые вернула модель для X_test
# y_test - это правильные ответы для X_test
print(accuracy_score(y_test,answers_pred))
```
