```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv('/Stars.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
train = df.drop(['Type'], axis=1) # удалим целевую переменную из тренировочных данных
print(train.isna().sum())
```

```python
# Преобразуем категориальные признаки (Color и Spectral_Class)
train = pd.get_dummies(train, columns=['Spectral_Class', 'Color'])
```

```python
train.head()
```

```python
from sklearn.model_selection import train_test_split

# train - на тренировочные и тестовые
X = train
y = df['Type']

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3)

model = LogisticRegression()
# model = KMeans()
# model = LinearRegression()

# модель учится, для этого ей нужны правильные
model.fit(X_train, y_train)

# тестируем, для признаков у которых нет ответов, мы предсказывает ответ
answers_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

# Как оценить насколько хорошо работает ваша модель?

# answers_pred - ответы которые вернула модель для X_test
# y_test - это правильные ответы для X_test
print(accuracy_score(y_test,answers_pred))
```
