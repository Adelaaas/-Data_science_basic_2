```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LogisticRegression
```

```python
train = pd.read_csv("train_weather.csv")
test = pd.read_csv("test_weather.csv")
```

```python
train.head(20)
```

```python
train.info()
```

```python
print(train.isna().sum())
```

```python
for i in train:
  if i != "Location" and i != 'WindDir9am' and i != 'WindDir3pm' and i != 'RainToday' and i != 'RainTomorrow' and i != 'WindGustDir':
    value = train[i].mean()
    train[i] = train[i].fillna(value)
train.head(20)
```

```python
print(train.shape)
print(train.isna().sum())
```

```python
train = train.drop(['WindGustDir'], axis=1)
train = train.drop(['WindDir9am'], axis=1)
train = train.drop(['WindDir3pm'], axis=1)
```

```python
train.head(20)
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
label_encoder.fit(train['Location']) 
train['Location'] = label_encoder.transform(train['Location']) 
train.head()
```

```python
from sklearn.model_selection import train_test_split

# train - на тренировочные и тестовые
X = train.drop(['RainTomorrow'], axis=1)
y = train['RainTomorrow']

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3)

model = LogisticRegression()
# model = KMeans()
# model = LinearRegression()

# модель учится, для этого ей нужны правильные
model.fit(X_train, y_train)

# тестируем, для признаков у которых нет ответов, мы предсказывает ответ
answers_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

# Как оценить насколько хорошо работает ваша модель?

# answers_pred - ответы которые вернула модель для X_test
# y_test - это правильные ответы для X_test
print(accuracy_score(y_test,answers_pred))
```
