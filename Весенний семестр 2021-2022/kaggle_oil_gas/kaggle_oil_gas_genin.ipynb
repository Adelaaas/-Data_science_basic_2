```python
tr=train.copy()
tr=tr['Tectonic regime'].str.get_dummies('/')
tr.columns=tr.columns.str.split().str[0]
tr=tr.add_prefix('Tectonic regime_').reset_index()
tr.drop(['index'],axis=1,inplace=True)
tr.shape[1]
```

```python
tr=pd.concat([train,tr],axis=1)
```

```python
tr1=tr.copy()
tr1=tr1['Structural setting'].str.get_dummies('/')
tr1.columns=tr1.columns.str.split().str[0]
tr1=tr1.add_prefix('Structural setting_').reset_index()
tr1.drop(['index'],axis=1,inplace=True)
tr1.shape[1]
```

```python
tr=pd.concat([tr,tr1],axis=1)
```

```python
tr2=tr.copy()
tr2=tr2['Period'].str.get_dummies('-')
tr2.columns=tr2.columns.str.split().str[0]
tr2=tr2.add_prefix('Period_').reset_index()
tr2.drop(['index'],axis=1,inplace=True)
tr2.shape[1]
```

```python
tr=pd.concat([tr,tr2],axis=1)
```

```python
tr=pd.get_dummies(tr,columns=['Hydrocarbon type','Reservoir status','Lithology'])
tr.drop(['Tectonic regime','Period','Structural setting'], axis=1, inplace=True)
tr
```

```python
from sklearn.preprocessing import LabelEncoder
lec=LabelEncoder()
tr['Onshore/Offshore'] = lec.fit_transform(tr['Onshore/Offshore'])
tr
```

```python
from imblearn.over_sampling import SMOTE
os=SMOTE(random_state=0,k_neighbors=3)
Xtr = tr.drop(['Onshore/Offshore'], axis=1)
Ytr = tr['Onshore/Offshore']
column = Xtr.columns
print("До балансировки")
print(Xtr.shape)
print(Ytr.value_counts())
osX,osY=os.fit_resample(Xtr,Ytr)
osX=pd.DataFrame(data=osX,columns=Xtr.columns)
osY=pd.DataFrame(data=osY,columns=['Onshore/Offshore'])
print("После балансировки")
print(osX.shape)
print(osY.value_counts())
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0,1))
scaler.fit(osX)
osXn=scaler.transform(osX)
osXn
```

```python
from sklearn.neighbors import KNeighborsClassifier
scores=[]
for i in range(1,10):
    knn=KNeighborsClassifier(i)
    knn.fit(osXn,osY)
    answers=knn.predict(osXn)
    scores.append(knn.score(osXn,osY))
print(scores)
```
