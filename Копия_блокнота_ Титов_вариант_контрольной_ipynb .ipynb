```python
# 3. Написать фукнцию, которая принимает DataFrame, название
# и возвращает кортеж ((максимальный возраст для Segmentation == А, минимальный возраст для Segmentation == А), (максимальный возраст для Segmentation == B, минимальный возраст для Segmentation == B))
# т е вывод: ((42, 10), (50, 11))

# ваш код тут

tup = ((df.loc[df['Segmentation'] == "А"]['Age'].max(), df.loc[df['Segmentation'] == "А"]['Age'].min()), (df.loc[df['Segmentation'] == "B"]['Age'].max(), df.loc[df['Segmentation'] == "B"]['Age'].min()))
print(tup)
```

```python
# 4. Построить график:
# - график зависимости Age от Family_Size
# - гистограмма распределения количества для каждого Spending_Score
# У графика должен быть заголовок, подписаны оси, а также легенда (т е графики должны быть подписаны)
import matplotlib.pyplot as plt
# ваш код тут
df.plot(x = 'Family_Size',y = 'Age', kind='scatter')
plt.title('Распределение возраста в зависимости от Family_Size')
plt.xlabel("Family_Size")        
plt.ylabel("Age") 
plt.show()

ax = df['Spending_Score'].value_counts().plot(kind='bar',
                                    figsize=(12,4),
                                    title="")
```
