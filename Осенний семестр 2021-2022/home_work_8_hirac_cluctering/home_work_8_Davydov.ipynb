```python
df = pd.read_csv("CC GENERAL.csv")
df
```

```python
print(df.shape) # размерность данных
print('______________________________________________________')
print(df.info()) # типы данных
print('______________________________________________________')
print(df.isna().sum()) # пустые значения
```

```python
df.dropna(subset=['MINIMUM_PAYMENTS'], inplace=True) # удаляем столбец с пустыми значениями
df
```

```python
df.columns
```

```python
X = df.loc[:,['PURCHASES', 'PAYMENTS']] # вектор признаков
y = df['BALANCE'] # вектор ответов
# BALANCE: сумма баланса, оставшаяся на их счету для совершения покупок.
# PURCHASES: количество покупок, совершенных со счета.
# PAYMENTS: сумма платежа, произведенного пользователем.
# Сортируем по признаку: Кто сколько тратит.
```

```python
 import matplotlib.pyplot as plt
 from sklearn.cluster import KMeans
 
 # метод локтя
 
 SSE = [] # Сумма квадратов ошибок для каждого результата

 for k in range(1,9):
     kmeans = KMeans(n_clusters = k) # строим кластер
     kmeans.fit(X)
     SSE.append(kmeans.inertia_)
 
 plt.plot(range(1,9), SSE, marker='s');
 plt.xlabel('k')
 plt.ylabel('SSE');
 plt.show()
```

```python
kmeans = KMeans(n_clusters=3) # взяли 3 кластера, т.к. там крутой изгиб
kmeans.fit(X)
y_pred = kmeans.labels_
df['cluster'] = y_pred
df
```

```python
plt.scatter(df['PURCHASES'],df['PAYMENTS']) # разбиваем объекты на кластеры
plt.xlabel('PURCHASES')
plt.ylabel('PAYMENTS')
plt.show()
```

```python
X['cluster'] = y_pred # создаём в df колонку, в которую записываем кластер
X
```

```python
plt.plot(X[X['cluster']==0]['PURCHASES'], X[X['cluster']==0]['PAYMENTS'], 'bo', label='class1')

plt.plot(X[X['cluster']==1]['PURCHASES'], X[X['cluster']==1]['PAYMENTS'], 'go', label='class2')

plt.plot(X[X['cluster']==2]['PURCHASES'], X[X['cluster']==2]['PAYMENTS'], 'ro', label='class3')

plt.xlabel('PURCHASES')
plt.ylabel('PAYMENTS')
plt.legend(loc=0)
plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering
```

```python
# матрица расстояний
Z = linkage(X, 'ward')

# визуализация дендрограмы
fig = plt.figure(figsize=(10, 10))
dn = dendrogram(Z)
```

```python
from sklearn.cluster import AgglomerativeClustering

hc = AgglomerativeClustering(n_clusters = 9, affinity = 'euclidean', linkage = 'ward')
# hc = AgglomerativeClustering(affinity = 'euclidean', linkage = 'ward', distance_threshold = 10)

y_hc = hc.fit_predict(X)
y_hc
```

```python
X['pred_sklearn'] = y_hc
X
```

```python
plt.scatter(X.loc[y_hc==0, 'PURCHASES'], X.loc[y_hc==0, 'PAYMENTS'], s=50, marker='o', color='red')
plt.scatter(X.loc[y_hc==1, 'PURCHASES'], X.loc[y_hc==1, 'PAYMENTS'], s=50, marker='o', color='blue')
plt.scatter(X.loc[y_hc==2, 'PURCHASES'], X.loc[y_hc==2, 'PAYMENTS'], s=50, marker='o', color='green')

plt.show()
```
