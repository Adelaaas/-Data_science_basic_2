```python
#N8
a = np.array([1.2, -10.3, 5.3, 6.4, -6.3, -3.4, 0], float)
for i in range(len(a)):
  if a[i]>0: 
    a[i]=np.ceil(a[i])
  elif a[i]<0: 
    a[i]=np.floor(a[i])
  else:
    a[i]=0
print(a)
```
